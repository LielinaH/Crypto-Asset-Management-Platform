```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load historical data
data_path = "../data/BTCUSDT_historical.csv"
btc_data = pd.read_csv(data_path, parse_dates=['open_time'])

# Plot closing price over time
plt.figure(figsize=(12, 6))
plt.plot(btc_data['open_time'], btc_data['close'], label='BTC Closing Price')
plt.title('BTC Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid()
plt.show()

# Calculate daily percentage change
btc_data['daily_return'] = btc_data['close'].pct_change() * 100
plt.figure(figsize=(12, 6))
plt.plot(btc_data['open_time'], btc_data['daily_return'], label='Daily Returns')
plt.title('BTC Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily % Change')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid()
plt.show()

# Rolling metrics
btc_data['rolling_mean'] = btc_data['close'].rolling(window=30).mean()
btc_data['rolling_std'] = btc_data['close'].rolling(window=30).std()

# Plot rolling metrics
plt.figure(figsize=(12, 6))
plt.plot(btc_data['open_time'], btc_data['close'], label='BTC Closing Price')
plt.plot(btc_data['open_time'], btc_data['rolling_mean'], label='30-Day Rolling Mean', linestyle='--')
plt.plot(btc_data['open_time'], btc_data['rolling_std'], label='30-Day Rolling Std Dev', linestyle='--')
plt.title('BTC Rolling Metrics')
plt.xlabel('Date')
plt.ylabel('Price/Std Dev')
plt.legend()
plt.grid()
plt.show()

# Statistical summary
stats_summary = {
    "Average Daily Return (%)": btc_data['daily_return'].mean(),
    "Max Daily Return (%)": btc_data['daily_return'].max(),
    "Min Daily Return (%)": btc_data['daily_return'].min(),
    "Standard Deviation of Daily Returns (%)": btc_data['daily_return'].std(),
}

print("Statistical Summary of BTC Daily Returns:")
for key, value in stats_summary.items():
    print(f"{key}: {value:.2f}")
```
